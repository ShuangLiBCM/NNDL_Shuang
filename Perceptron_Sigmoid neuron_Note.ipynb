##  Creat Sigmoid Function 

```python
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1/(1+ np.exp(-z))
```

```python
a = np.linspace(-5,5,100);
b = sigmoid(a);
plt.plot(a,b,'b.-')
plt.plot(a,sigmoid(100*a),'r.-')
plt.ylabel('$\sigma(x)$')
plt.xlabel('$x$')
```

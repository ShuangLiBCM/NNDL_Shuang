## Perceptron and related algorithms

The earliest algorithm to learn from the input pattern and generate the weights for predicting was inspired from McCullock-Pitts (MCP) neuron. The idea was to learn based on the input pattern automatically the weights to predict if a neuron will fire or not. In the perceptron algorithm, an activation function is defined as a threshold function over the weighted sum of the inputs.


$z = w_{1}x_{1} + w_{2}x_{2}+... +w_{n}x_{n} $

And the nonlinearty if defined as a heaviside step function

$ \phi(z) = 1, if z \geq \theta$
$ \phi(z) = -1, if z< \theta$

The basic idea of an activaiton function is to squash the net inputs into a binary output.

The training process could be seperated into 3 steps:
1. Initialze the weights into 0 or small random number
2. Calcualted the predicted binary number
3. Based on the relationship with the label, update the weights

The weights is updated based on the Rosenblatt's perceptron learning rule:

$\Delta w_{j} = \eta(y^{i}-y_{predict}^{i})x_{j}^{(i)}$

$\eta $ is the learning rate, which is a constant between 0 and 1

The learning rule will only converge if the two classes of data is linearly separabale.As could be seen from the description, perceptron is only one layer of network and is guided through the error comparing with the labeled data.

```python
import numpy as np
class Perceptron(object):
    """perceptron classifier
    Parameter:
    eta: float, learning rate (between 0.0 - 1.0)
    n_iter: int, passes over the training dataset
    
    Attributes:
    w_: 1d-array, weights after fitting
    error_: list, number of misclassification in every epoch
    """
    def __init__(self, eta = 0.1,n_iter = 10):
        self.eta = eta
        self.n_iter = n_iter
        
    def fit(self,X,y):
        """fitting training data
        Parameter:
        X: {array-like}, shape = [n_samples,n_features]
        y: array-like,shape = [n_samples, 1]
        Returns: self:object
        """
        self.w_ = np.zeros(1 + X.shape[1])
        self.errors_ = []
        
        for _ in range(self.n_iter):
            errors = 0
            for xi,target in zip(X,y):   # elementwise computation for all samples, different from follwing batch update
                update = self.eta * (target - self.predict(xi))   # Part of the Rosenlatt's perceptron learning rule
                self.w_[1:] += update*xi   # Update weights
                self.w_[0] += update    # Update bias
                errors += int(update != 0.0)   # Get the number os misclassificationi
                # indentation here is critical
            self.errors_.append(errors)
        return self
        
    def net_input(self,X):
        return np.dot(X,self.w_[1:])+self.w_[0]
    
    def predict(self,X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)   # Useful function from numpy
```

Now we could generate the decision boundary. 

```python
from matplotlib.colors import ListedColormap

def plot_decision_region(X,y,classifier,resolution = 0.02):
    
    markers = ('s','x','o','^','v')
    colors = ('red','green','lightgreen','gray','cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min,x1_max = X[:,0].min()-1,X[:,0].max()+1
    x2_min,x2_max = X[:,1].min()-1,X[:,1].max()+1
    xx1,xx2 = np.meshgrid(np.arange(x1_min,x1_max,resolution),np.arange(x2_min,x2_max,resolution))
    Z = classifier.predict(np.array([xx1.ravel(),xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1,xx2,Z,alpha = 0.4,cmap = cmap)
    plt.xlim(xx1.min(),xx1.max())
    plt.ylim(xx2.min(),xx2.max())
    
    # Plot class labels
    for idx, c1 in enumerate(np.unique(y)):
        plt.scatter(x = X[y==c1,0],y=X[y==c1,1],alpha = 0.8, c = cmap(idx),marker = markers[idx],label = c1)
    
```

```python
plot_decision_region(X,y,classifier = ppn)
plt.xlabel('septal len (cm)')
plt.ylabel('petal len (cm)')
plt.legend(loc = 'upper left')
```

Summary: perceptron could be used to learn a classifier for seperating linearly seperable group. However, Frank Rosenblatt proofed that if the bondary hyperplane is not linear, the weights will keep updating until the maximum number of the training limit achieves. 

## Adaptive linear neuron

Adaptive linear neuron algorithm (Adaline) brings up the key concept about minimizing the cost function as a goal of learing. The key differences between the Adaline and Rosenblatt's algorithm is that the error signal (learning) is performed based on a linear activation function instead of a unit step function. 


$\phi(w^{T}x)=w^{T}x$

A unit step function is added on top of the output of the liear activation function for predicting the class label. But is not used for updating the weights.

```python
Image(filename='Adaline algorithm.png')
```

One key idea in supervised learing is to update the weights according to a learning rule that minimize the cost function. In Adaline algorithm, the cost function is defined as the Sum of Squared Errors (SSE) of the calculated outcome and the true class label.

$J(w)= \frac12\sum_{i} [y^{i}-\phi(z^{i})]^2$  The cost function is defined over a range of training samples. Therefore, definition of batch size is important.

The key advantage of the cost function is that it continuous, instead of the step funciton, that is differentiable. The second advantage is that it is convex. Therefore, there exist an optimization solution.The method that we use to find the optimization solution is called gradient descent.

The updating rule in the cost function and gradient decay scenario becomes:

$w: = w+\Delta w$

$\Delta w = -\eta \Delta J(w)$ update the weights towards the negative gradients with learning rate influencing the step size

$\Delta J(w_{j}) = \frac {\partial J} {\partial w_{j}} = \sum_{i}[y^{i}-\phi(z^{i})](-x_{j}^{i})$

$\Delta w_{j} = \eta \sum_{i}[y^{i}-\phi(z^{i})]x_{j}^{i}$

The weight is updated after summarizing all the input sample. Therefore is also called "batch" gradient descent.

```python
class AdalineGD(object):
    """
    parameters:eta, n_iter;
    attributes: w_, errors_
    """
    
    def __init__(self,eta = 0.01, n_iter = 50):
        self.eta = eta
        self.n_iter = n_iter
        
    def fit(self,X,y):
        self.w_ = np.zeros(1+X.shape[1])
        self.cost_ = []
        
        for i in range(self.n_iter):
            output = self.net_input(X)
            errors = y - output
            self.w_[1:] += self.eta*X.T.dot(errors)    # Batch learning
            self.w_[0] += self.eta*errors.sum()
            cost = (errors**2).sum()/2.0
            self.cost_.append(cost)
            
    def net_input(self,X):
        return np.dot(X,self.w_[1:])+self.w_[0]
    
    def activation(self,X):
        return self.net_input(X)
    
    def predict(self,X):
        return np.where(self.activation(X) >=0.0, 1, -1)    # Predict the label
```

```python
fig,ax = plt.subplots(nrows = 1,ncols = 2, figsize = (8,4))
ada1 = AdalineGD(n_iter = 10, eta = 0.01)
ada1.fit(X,y)
ax[0].plot(range(1,len(ada1.cost_)+1),np.log10(ada1.cost_),marker = 'o')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('log of sum cost')
ax[0].set_title('Adaline, learning rate =  0.01')
ada2 = AdalineGD(n_iter = 1000,eta = 0.0001)
ada2.fit(X,y)
ax[1].plot(range(1,len(ada2.cost_)+1),np.log10(ada2.cost_),marker = 'o')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('log of sum cost')
ax[1].set_title('Adaline, learning rate =  0.001')
```

One possible reason that leads to sensitivity to hyperparameter is the non-standadize data scale. To eliminate it, data standadization is always conducted before training was applied.

```python
X_std = np.copy(X)
X_std[:,0] = (X[:,0]-X[:,0].mean())/X[:,0].std()
X_std[:,1] = (X[:,1]-X[:,1].mean())/X[:,1].std()

fig,ax = plt.subplots(nrows = 1,ncols = 2, figsize = (8,4), sharey=True)
ada1.fit(X_std,y)
ax[0].plot(range(1,len(ada1.cost_)+1),ada1.cost_,marker = 'o')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('log of sum cost')
ax[0].set_title('Adaline, learning rate =  0.01')
ada2 = AdalineGD(n_iter = 1000,eta = 0.0001)
ada2.fit(X_std,y)
ax[1].plot(range(1,len(ada2.cost_)+1),ada2.cost_,marker = 'o')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Sum cost')
ax[1].set_title('Adaline, learning rate =  0.001')
```

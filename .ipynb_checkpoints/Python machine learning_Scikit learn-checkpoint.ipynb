## Using Scikitlearn library for machine learning

We have 3 goals in the learning here. 1. Basic concepts of popular classification algorithms.2. How to use scikit-learn library. 3. How to choose a machine learning algorithm.

There are several factors that we need to consider when choosing a learning algorithm. Popular ones are num of features/samples, the amount of noise and whether they are linearly seperable or not.

The five key steps that involve training a machine learning task involve: 
1. Select feature;
2. Choosing a performace matric;
3. Choosing a classifier/learning algorithm;
4. Evaluating the performance;
5. Tuning


```python
from sklearn import datasets
import numpy as np

iris = datasets.load_iris()
X = iris.data[:,[2,3]]
y = iris.target

# Use sklearn library to generate training and testing set
from sklearn.cross_validation import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 0)

# Use sklearn library to do feature scaling

from sklearn.preprocessing import StandardScaler
# StandardScaler is a class, instantiation is needed
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)  # Use same mean and std to standardize both tranining and testing set

# Implement perceptron with sklean library
from sklearn.linear_model import Perceptron
ppn = Perceptron(n_iter = 40,eta0 = 0.1,random_state = 0)
ppn.fit(X_train_std,y_train)     # Perceptro could be used to train multi classes, each class has a perceptron classifier

# Use the testset to generate labels 
y_predict = ppn.predict(X_test_std)

# Get the percentage of correction
print(1-(y_test!=y_predict).sum()/len(y_test))
```

```python
# Visualize the decision boudary and classification result
from matplotlib.colors import ListedColormap
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def plot_decision_region(X,y,classifier,test_idx,resolution = 0.02):
    
    markers = ('s','x','o','^','v')
    colors = ('red','blue','lightgreen','gray','cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min,x1_max = X[:,0].min()-1,X[:,0].max()+1
    x2_min,x2_max = X[:,1].min()-1,X[:,1].max()+1
    xx1,xx2 = np.meshgrid(np.arange(x1_min,x1_max,resolution),np.arange(x2_min,x2_max,resolution))
    Z = classifier.predict(np.array([xx1.ravel(),xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1,xx2,Z,alpha = 0.4,cmap = cmap)   # filled contour plot
    plt.xlim(xx1.min(),xx1.max())
    plt.ylim(xx2.min(),xx2.max())
    
    # Plot all samples
    for idx, c1 in enumerate(np.unique(y)):
        plt.scatter(x = X[y==c1,0],y=X[y==c1,1],alpha = 0.8, c = cmap(idx),marker = markers[idx],label = c1)
    
    if test_idx:
        X_test,y_test = X[test_idx,:],y[test_idx]
        plt.scatter(X_test[:,0],X_test[:,1],alpha = 1.0, c = '',linewidth = 1,marker = 'o',label = 'test_set')
```

```python
X_combined_std = np.vstack((X_train_std,X_test_std))
y_combined = np.hstack((y_train,y_test))
plot_decision_region(X = X_combined_std,y = y_combined,classifier = ppn,test_idx = range(105,150))
plt.xlabel('septal len standadized')
plt.ylabel('petal len standadized')
plt.legend(loc = 'upper left')
```

As discucssed before, a perceptron learning algorithm never converges on a dataset that is not linearly seperatble. To use a better classificaiton algorithm that converges even the data is not linearly seperable. We will use logistic regression.

## Logistic regression with scikitlearn

Logistic regression is easy to think as one way to convert the real numbers to probability.A straight forward activation function could be sigmoid function. 

$\phi(x) = \frac1 {1+e^{-z}} $

,where z is the net input, $z = w^{T}x$

```python
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline

def sigmoid(z):
    return (1.0/(1.0+np.exp(-z)))

z = np.arange(-7,7,0.1)
phi_z = sigmoid(z)

plt.plot(z,phi_z)
plt.axvline(0.0,color = 'k')
plt.axhspan(0.0, 1.0, facecolor='1.0', alpha=1.0, ls='dotted')
plt.axhline(y=0.5, ls='dotted', color='k')
plt.yticks([0.0, 0.5, 1.0])
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\phi (z)$')
```

Logistic regression could be understood similar to Adaline classifier.The difference is that the activation function is changed from an identification function to Sigmoid function.The output of the Sigmoid function could be understood as the probability of a sample belong to a perticular class, given its feature x and the trained weights. In this case, we could classify the sample into the category where it shows high probabiltiy.

In adaline classifier, we develop the cost function as the sum of squared errors for each traiing sample and conduct gradiend descent. The cost function in logistic regression is derived from likelihood function. Assuming that we obtained the sample through identical independent distribution, the likelihood funciton then becomes:

$L(w) = P(y|x,w) = \prod_{i} P(y^{i}|x^{i},w) = \prod_{i} \phi(x^{i},w)= \prod_{i}\phi(x^{i},w)^{y^{i}}[1-\phi(x^{i},w)]^{1-y^{i}} $

In pratice, it is easiear to maximize the natural log of the likilihood function, as can be seen:
$L(w) = log P(y|x,w) = \sum {y^{i}log\phi(x^{i},w)+(1-y^{i})log[1-\phi(x^{i},w)]}$

There are two major reasons for applying loglikelihood. One is for converting the product function into sum. The other is to prevent numerical underflow, which is possible calculating the probabilities.

Now we could use a gradient descent algorithm to maximize the likelihood function, or minimize the negative log-likelihood. We could call it the cost function.

$J(w) = -log P(y|x,w) = \sum -{y^{i}log\phi(x^{i},w)+(1-y^{i})log[1-\phi(x^{i},w)]}$
